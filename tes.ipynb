```python
# Importing the library
import pandas as pd
import plotly.express as px
import plotly.offline as pyo

pyo.init_notebook_mode(connected=True)

# Reading the data
data = pd.read_csv('data.csv')
print(data)

# Sorting the DataFrame by specification rating
df_wr = data.sort_values('spec_rating')

# Creating a bar chart
fig = px.bar(
    df_wr,
    x='brand',
    y='spec_rating',
    color='processor',
    labels={"processor": "processor"},
    title='The Distribution of Specification Ratings',
    hover_data=['processor']
)

fig.show()
```
